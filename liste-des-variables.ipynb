# Liste des variables OpenFisca

Ce notebook montre comment extraire la liste des variables d'entrée et de sortie, pour les individus et les entités.

```python
import collections
import json
import urllib2
```

```python
try:
    response = urllib2.urlopen('http://api.openfisca.fr/api/1/fields')
except urllib2.HTTPError as response:
    print response.read()
    raise
response_json = json.loads(response.read(), object_pairs_hook = collections.OrderedDict)
```

## Variables en entrée

### Variables en entrée sur les individus

```python
for name, variable in response_json['columns'].iteritems():
    if variable['entity'] == 'ind':
        label = variable['label']
        if len(label) > 90:
            label = label[:87] + '...'
        print u'{:24s} | {:11} | {:90s}'.format(variable['name'], variable['@type'], label)
        if variable['@type'] == 'Enumeration' and variable.get('labels'):
            for index, option in variable['labels'].iteritems():
                print u'{:24s} | {:11} | * {:2} = {}'.format(u'', u'', index, option)
```
